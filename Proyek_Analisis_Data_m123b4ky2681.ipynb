## Visualization & Explanatory Analysis

### Pertanyaan 1:

````python
data_harian = hour_df.groupby('weekday')['count'].sum().reset_index()

label_harian = {
    0: 'Minggu',
    1: 'Senin',
    2: 'Selasa',
    3: 'Rabu',
    4: 'Kamis',
    5: 'Jumat',
    6: 'Sabtu'
}

# Map the weekday numbers to their corresponding labels
data_harian['weekday'] = data_harian['weekday'].map(label_harian)

# Sort based on the original weekday order (0-6)
order_harian = ['Minggu', 'Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu']

plt.figure(figsize=(10, 6))
sns.barplot(x='weekday',
            y='count',
            data=data_harian,
            order=order_harian)  # Ensures the correct order of days
plt.title('Jumlah Penyewaan Sepeda Harian')
plt.xlabel('Hari')
plt.ylabel('Jumlah Penyewaan')
plt.show()
````

### Pertanyaan 2:

````python
data_perjam = hour_df.groupby('hour')['count'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='hour',
             y='count',
             data=data_perjam,
             marker='o',
             color='green')
plt.title('Jumlah Penyewaan Sepeda Per Jam')
plt.xlabel('Jam')
plt.ylabel('Jumlah Penyewaan')
plt.xticks(range(0, 24))
plt.grid(True)
plt.tight_layout()
plt.show()
````
